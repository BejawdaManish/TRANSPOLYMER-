```python
def compute_features(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if not mol:
        return [0]*10 + [0]*2048
    desc = [fn(mol) for fn in descriptor_fns]
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
    return desc + list(fp)

features = dataset["Aug_SMILES"].apply(compute_features).tolist()
features_tensor = torch.tensor(features, dtype=torch.float32)

# ðŸ¤– EMBEDDINGS FROM ChemBERTa
tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
chemberta = AutoModel.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
chemberta.eval()

@torch.no_grad()
def embed_smiles(smiles_list):
    inputs = tokenizer(smiles_list, return_tensors="pt", padding=True, truncation=True, max_length=128)
    outputs = chemberta(**inputs)
    return outputs.last_hidden_state[:, 0, :]  # CLS token

cls_embeddings = []
batch_size = 32
for i in range(0, len(dataset), batch_size):
    smiles_batch = dataset["Aug_SMILES"].iloc[i:i+batch_size].tolist()
    emb = embed_smiles(smiles_batch)
    cls_embeddings.append(emb)
cls_tensor = torch.cat(cls_embeddings, dim=0)

# ðŸŽ¯ TARGETS
target_cols = ["Tensile_strength(Mpa)", "Ionization_Energy(eV)", "Electron_Affinity(eV)", "LogP", "Refractive_Index", "Molecular_Weight(g/mol)"]
targets = dataset[target_cols]
scalers = [StandardScaler().fit(targets[[col]]) for col in target_cols]
y_scaled = np.hstack([scaler.transform(targets[[col]]) for scaler, col in zip(scalers, target_cols)])
y_tensor = torch.tensor(y_scaled, dtype=torch.float32)

# ðŸ”€ SPLIT DATA
X_train_val, X_test, y_train_val, y_test, f_train_val, f_test = train_test_split(cls_tensor, y_tensor, features_tensor, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val, f_train, f_val = train_test_split(X_train_val, y_train_val, f_train_val, test_size=0.2, random_state=42)

train_loader = DataLoader(TensorDataset(X_train, f_train, y_train), batch_size=16, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val, f_val, y_val), batch_size=16)
test_loader = DataLoader(TensorDataset(X_test, f_test, y_test), batch_size=16)

# model.py

import torch.nn as nn
from torch.nn import TransformerEncoder, TransformerEncoderLayer

class TransformerRegressor(nn.Module):
    def __init__(self, emb_dim=768, feat_dim=2058, output_dim=6, nhead=8, num_layers=2):
        super().__init__()
        self.feat_proj = nn.Linear(feat_dim, emb_dim)
        encoder_layer = TransformerEncoderLayer(d_model=768, nhead=8, dim_feedforward=1024, dropout=0.1, batch_first=True)
        self.transformer_encoder = TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.regression_head = nn.Sequential(
            nn.Linear(emb_dim, 256), nn.ReLU(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, output_dim)
        )

    def forward(self, x, feat):
        feat_emb = self.feat_proj(feat)
        stacked = torch.stack([x, feat_emb], dim=1)
        encoded = self.transformer_encoder(stacked)
        aggregated = encoded.mean(dim=1)
        return self.regression_head(aggregated)


# ðŸ‘¥ ENSEMBLE
num_models = 3
dev = torch.device("cuda" if torch.cuda.is_available() else "cpu")
preds_all = []

for seed in range(num_models):
    torch.manual_seed(seed)
    model = TransformerRegressor().to(dev)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.5)
    criterion = nn.SmoothL1Loss(reduction='none')
    target_weights = torch.tensor(1 / np.std(y_scaled, axis=0), dtype=torch.float32).to(dev)

    best_val_loss = float('inf')
    patience, counter = 5, 0

    for epoch in range(100):
        model.train()
        for xb, fb, yb in train_loader:
            xb, fb, yb = xb.to(dev), fb.to(dev), yb.to(dev)
            optimizer.zero_grad()
            preds = model(xb, fb)
            loss = criterion(preds, yb)
            weighted = (loss * target_weights).mean()
            weighted.backward()
            optimizer.step()

        # Validation
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for xb, fb, yb in val_loader:
                preds = model(xb.to(dev), fb.to(dev))
                loss = criterion(preds, yb.to(dev))
                val_loss += (loss * target_weights).mean().item()
        val_loss /= len(val_loader)
        print(f"Epoch {epoch+1} | Validation Loss: {val_loss:.4f}")
        scheduler.step(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), f"best_model_seed{seed}.pt")
            counter = 0
        else:
            counter += 1
            if counter >= patience:
                break

    # Load best model & Predict
    model.load_state_dict(torch.load(f"best_model_seed{seed}.pt"))
    model.eval()
    preds = []
    with torch.no_grad():
        for xb, fb, _ in test_loader:
            preds.append(model(xb.to(dev), fb.to(dev)).cpu())
    preds_all.append(torch.cat(preds))

# ðŸ§ª Final ensemble prediction
y_preds_ens = torch.stack(preds_all).mean(dim=0).numpy()
y_true = y_test.numpy()
y_pred = np.column_stack([scaler.inverse_transform(y_preds_ens[:, i:i+1]) for i, scaler in enumerate(scalers)])
y_true = np.column_stack([scaler.inverse_transform(y_true[:, i:i+1]) for i, scaler in enumerate(scalers)])

# ðŸ“Š METRICS
print("\nðŸ“Š Final Ensemble Test Results:")
print(f"MAE: {mean_absolute_error(y_true, y_pred):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_true, y_pred)):.4f}")
print(f"RÂ²: {r2_score(y_true, y_pred):.4f}\n")

# ðŸ”¬ Per-Target Results
target_names = ["Tensile Strength (MPa)", "Ionization Energy (eV)", "Electron Affinity (eV)", "LogP", "Refractive Index", "Molecular Weight (g/mol)"]
for i, name in enumerate(target_names):
    mae = mean_absolute_error(y_true[:, i], y_pred[:, i])
    rmse = np.sqrt(mean_squared_error(y_true[:, i], y_pred[:, i]))
    r2 = r2_score(y_true[:, i], y_pred[:, i])
    print(f"{name}: MAE = {mae:.4f}, RMSE = {rmse:.4f}, RÂ² = {r2:.4f}")
```
